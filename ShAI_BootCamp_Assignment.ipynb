# Data Cleaning
Handle missing data by suitable method with explain why you use it.

##### dropping 'Notes' and 'Status' columns becuase they're 100% missing values

```python
new_df = df.drop(columns=['Notes', 'Status'])
```

##### imputing 'Benefits' and 'BasePay' values because they represent about 25% and 0.5% respectively of the whole data

```python
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

numeric_cols = ['Benefits','BasePay']

preprocessor = ColumnTransformer(
    transformers=[
        ('numeric', SimpleImputer(strategy='mean'), numeric_cols),
    ])

data_imputed = pd.DataFrame(preprocessor.fit_transform(new_df))
```

```python
new_df[['Benefits','BasePay']] = data_imputed
```

##### dropping empty rows for the rest because they represent < 0.003

```python
new_df = new_df.dropna()
```

```python
new_missing = new_df.isnull().sum()

print(new_missing)
```

# Basic Data Visualization
Create histograms or bar charts to visualize the distribution of salaries, and use pie charts to represent the proportion of employees in different departments.

##### I couldn't find 'Department' column so I created  a pie chart of the proportion of employees in n most recrurrent titles

# Grouped Analysis
Group the data by one or more columns and calculate summary statistics for each group, and compare the average salaries across different groups.

```python
grouped = df.groupby('JobTitle')['TotalPay'].max()
sorted_grouped = grouped.sort_values(ascending=False)

print(sorted_grouped)
```
